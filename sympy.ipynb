```python
import sympy as sp
```

```python
x, y = sp.symbols("x y")

x, y
```

```python
f = x**2 + y**2  # определяем функцию f(x, y
f
```

```python
f.diff(x) # вычисляем производную функции f по x
```

```python
x_int = f.integrate(x)
y_int = f.integrate(y)  # вычисляем интеграл функции f по y
x_int, y_int  # выводим результаты интегрирования
```

```python
sp.solve(f, (x, y))  # решаем уравнение f(x, y) = 0 для x=1 и y=2
```

```python
f.subs({x: 10, y: 2})  # подставляем значения x=1 и y=2 в функцию f
```

```python
sp.plot(f.subs(y, 0), (x, -10, 10))  # строим график функции f(x, y=0) в диапазоне x от -10 до 10
```

```python
import numpy as np

# Преобразуем в функцию, которую можно вычислить численно
f_lamb = sp.lambdify((x, y), f, "numpy")
f_lamb
```

```python
x_vals = np.linspace(-50, 50, 100)
y_vals = np.linspace(-50, 50, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f_lamb(X, Y)
# создали дохера точек для этой сетки и положили из в Z
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
```
