```python
import matplotlib.image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import math

# Функция для сжатия одного канала
def compress_channel(channel, k):
    U, Sigma, Vt = np.linalg.svd(channel, full_matrices=False)  # full_matrices=False для экономии памяти
    U_k = U[:, :k]
    Sigma_k = np.diag(Sigma[:k])
    Vt_k = Vt[:k, :]
    return np.dot(U_k, np.dot(Sigma_k, Vt_k))

# Сжимаем каждый канал для заданного k
def compress_photo(photo, k):
    compressed = np.zeros_like(photo)
    for i in range(3):  # Для красного, зелёного, синего
        compressed[:,:,i] = compress_channel(photo[:,:,i], k)
    compressed = np.clip(compressed, 0, 255)  # Обрезаем значения в 0-255
    return compressed

# Определяем k
k_values = [500, 400, 300, 200, 10]

# Визуализация
plt.figure(figsize=(12, 4))

# Исходная картинка
plt.subplot(2, math.ceil((len(k_values) +1)/2), 1)
plt.imshow(photo)
plt.title(f'Исходная картинка, k={len(photo)}, Ошибка=0.00')
plt.axis('off')

# Сжимаем и показываем для каждого k
for i, k in enumerate(k_values, 1):
    photo_compressed = compress_photo(photo, k)
    matplotlib.image.imsave(f'compressed_photo_k_{k}.jpg', photo_compressed.astype(np.uint8))
    print(photo_compressed.size)
    
    # Считаем ошибку (MSE)
    mse = np.mean((photo - photo_compressed)**2)
    
    plt.subplot(2, math.ceil((len(k_values) +1)/2), i + 1)
    plt.imshow(photo_compressed)
    plt.title(f'Сжатая, k={k}, Ошибка={mse:.2f}')
    plt.axis('off')

plt.show()
```

```python
# Импортируем нужные библиотеки
import numpy as np
import matplotlib.pyplot as plt

# Создаём случайную таблицу 10 на 10 с числами от 0 до 255
np.random.seed(42)  # Для одинаковых результатов
image = np.random.randint(0, 256, size=(10, 10))

# Делаем SVD разложение
U, Sigma, Vt = np.linalg.svd(image, full_matrices=True)

# Сжимаем с k=5
k = 5
U_k = U[:, :k]
Sigma_k = np.diag(Sigma[:k])
Vt_k = Vt[:k, :]
image_compressed_5 = np.dot(U_k, np.dot(Sigma_k, Vt_k))

# Сжимаем с k=10
k = 10
U_k = U[:, :k]
Sigma_k = np.diag(Sigma[:k])
Vt_k = Vt[:k, :]
image_compressed_10 = np.dot(U_k, np.dot(Sigma_k, Vt_k))

# Считаем ошибку (среднеквадратичное отклонение)
mse_5 = np.mean((image - image_compressed_5)**2)
mse_10 = np.mean((image - image_compressed_10)**2)

# Показываем результаты
plt.figure(figsize=(12, 4))

# Исходная картинка
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Исходная картинка')
plt.axis('off')

# Сжатая с k=5
plt.subplot(1, 3, 2)
plt.imshow(image_compressed_5, cmap='gray')
plt.title('Сжатая, k=5')
plt.axis('off')

# Сжатая с k=10
plt.subplot(1, 3, 3)
plt.imshow(image_compressed_10, cmap='gray')
plt.title('Сжатая, k=10')
plt.axis('off')

plt.show()

# Выводим ошибки
print('Ошибка для k=5:', mse_5)
print('Ошибка для k=10:', mse_10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)
X_svd = np.dot(U[:, :2], np.diag(S[:2]))
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=iris.target, cmap='viridis')
plt.xlabel('Первая компонента')
plt.ylabel('Вторая компонента')
plt.title('SVD на данных Iris')
plt.show()

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target, cmap='viridis')
plt.xlabel('Первая компонента')
plt.ylabel('Вторая компонента')
plt.title('PCA на данных Iris')
plt.show()
```
