```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url, header=None)
scaler_std = StandardScaler()
scaler_minmax = MinMaxScaler()
std_scaled = scaler_std.fit_transform(iris.iloc[:, :-1])
minmax_scaled = scaler_minmax.fit_transform(iris.iloc[:, :-1])
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(std_scaled[:, 0], bins=20, color='blue')
plt.title("StandardScaler")
plt.subplot(1, 2, 2)
plt.hist(minmax_scaled[:, 0], bins=20, color='orange')
plt.title("MinMaxScaler")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

#   ( train.csv  https://www.kaggle.com/c/digit-recognizer/data)
data = pd.read_csv('./titanic/train.csv')  #    

#    
pre_x = data.drop(['Name', 'Cabin', 'Ticket'], axis=1)
pre_x['Sex'] = pre_x['Sex'].agg(lambda x: x == 'male')
pre_x['Embarked'] = pre_x['Embarked'].agg(lambda x: 1 if x == 'S' else (2 if x == 'C' else 3))
pre_x.dropna(inplace=True)

X = pre_x.drop('Survived', axis=1).values
y = pre_x['Survived'].values

#     (, 1000 )
n_samples = 1000
X_sample = X[:n_samples]
y_sample = y[:n_samples]
```

```python
#  t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_sample)
print(X_tsne)

# 
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_sample, cmap='tab10')
plt.colorbar(scatter, label='')
plt.title('t-SNE  MNIST (1000 )')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
iris = load_iris()
X = iris.data
y = iris.target
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.xlabel(" ")
plt.ylabel(" ")
plt.title("t-SNE  Iris")
plt.show()
```

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data
y = mnist.target.astype(int)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X[:1000])  #   
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X[:1000])
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y[:1000], cmap='tab10')
plt.title("PCA  MNIST")
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y[:1000], cmap='tab10')
plt.title("t-SNE  MNIST")
plt.show()
```
