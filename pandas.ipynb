```python
import pandas as pd
import numpy as np
```

```python
data = np.ndarray(1000).reshape(100, 10)
data
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.iat[1, 1]
```

```python
class Summator:
    start = 0

    def sum(self, x):
        self.start += x

    def result(self):
        return self.start


summator = Summator()

df[1].agg(summator.sum)

summator.result()
```

```python
import matplotlib.pyplot as plt
df.cumsum()
plt.figure()
df.plot()
plt.legend(loc='best')
```
